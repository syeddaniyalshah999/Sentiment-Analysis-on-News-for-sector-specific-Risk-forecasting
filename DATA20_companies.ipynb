```python
import yfinance as yf
import pandas as pd
```

```python
tickers = tickers = ['FSLR','VWDRY','VWSYF','ENPH','RUN','PLUG','SEDG','DQ','JKS','AMRC',
'ARRY','REX','CSIQ','BLDP','GPRE','NLLSF','GEVO','ADN','ASTI','AWSL','BNRG','CLNV','CREG',
'FHLD','GSFI','HTOO','MMMW','NEWH','OPVS','PRPS','TPICQ','ALTO','FCEL','TOYO','FF','HYSR',
'IMPP','IMPPP','AMTX','STEM','OPTT','AACTF','MAXN','BIOF','BEEM','BWEN','ZEO','AFPW','CCGY',
'CGEI','CPWR','FCELB','GCHK','GERS','HOKUQ','HTOOW','MEIL','NTCXF','SIRC','SOEN','SPIEF','SUNWQ']
```

```python
data = yf.download(tickers, start = '2017-01-01', end = '2023-12-30')
```

```python
close_volume = data.loc[:, ["Close", "Volume"]]
```

```python
close_volume_flat = close_volume.copy()
close_volume_flat.columns = [f"{ticker}_{price}" for price, ticker in close_volume_flat.columns]
```

```python
close_volume = data.loc[:, ["Close", "Volume"]]

close_volume = close_volume.swaplevel(axis=1)

close_volume = close_volume.sort_index(axis=1, level=0)

close_volume.columns = [f"{ticker}_{price}" for ticker, price in close_volume.columns]
```

```python
close_volume.to_csv("close_volume_companies2017to2023.csv")
```
