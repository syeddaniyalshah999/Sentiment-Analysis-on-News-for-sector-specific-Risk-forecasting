```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
# 1. Define Tickers
# "GC=F" = Gold Futures, "EUR=X" = USD to EUR exchange rate
tickers = ["GC=F", "EUR=X"]
```

```python
# 2. Set the specific date range
start_date = "2017-01-01"
end_date = "2023-12-31"
```

```python
# 3. Download the data
# The library fetches Open, High, Low, Close, Adj Close, and Volume by default
data = yf.download(tickers, start=start_date, end=end_date)['Close']
```

```python
# 4. Clean and Prepare
# Rename columns for clarity
data.columns = ['USD_EUR_Ratio', 'Gold_Price']
```

```python
# Remove rows with missing values (market holidays often differ)
data = data.dropna()

# 4. Save and Preview
data.to_csv("gold_usd_eur_regression_data.csv")
print(data.head())
```
