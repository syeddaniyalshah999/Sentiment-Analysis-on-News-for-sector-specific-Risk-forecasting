```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("Downloads/sp500_headlines_2008_2024.csv")   # change file name if different
```

```python
# Initialize sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
# Function to get sentiment score
def get_sentiment(text):
    if isinstance(text, str):  # avoid NaN or None
        return analyzer.polarity_scores(text)['compound']
    else:
        return 0
```

```python
# Apply sentiment function
df['sentiment_score'] = df['Headlines'].apply(get_sentiment)
```

```python
# Label sentiment
df['sentiment_label'] = df['sentiment_score'].apply(
    lambda x: 'Positive' if x > 0.05 else ('Negative' if x < -0.05 else 'Neutral')
)
```

```python
# Save results
df.to_csv("news_with_sentiment.csv", index=False)
```
